### Evaluación de los modelos LLM

Este notebook tiene como objetivo realizar una comparación entre distintos modelos de lenguaje ejecutados localmente, utilizando el mismo prompt de entrada para todos ellos. Se evaluarán cinco modelos: LLama3.1, Mistral7b, Phi3, TinyLlama y Gpt2.

Ejemplo de lista generada para comparar los modelos de lenguaje.

```python
detected = ['orange', 'apple', 'tomato', 'apple', 'tomato', 'banana', 'banana', 'tomato', 'tomato', 'cucumber', 'carrot']
unique_list = ['banana', 'apple', 'carrot', 'cucumber', 'tomato', 'orange']
```

Se inicializa el cliente de oLlama.

```python
import ollama

#Initialize ollama client

client = ollama.Client(host="http://172.24.32.1:11434")

def generar_receta_y_categorias(ingredients,model):
    prompt = f"""
        Tengo esta lista de alimentos detectados:

        {ingredients}

        1. Clasifícalos en categorías culinarias: Verduras, Frutas, Lácteos, Carnes/Pescados, Bebidas, Otros.
        

        2. Con los ingredientes disponibles, crea una receta completa con un Título, Descripción, Ingredientes (Solo existentes en la lista), Pasos y Sugerencias.
    

        Devuélvelo en formato claro y ordenado.
        """

    response = client.chat(
        model=model,
        messages=[{"role": "user", "content": prompt}]
    )
    return response['message']['content']

ingredients = unique_list
```

#### Llama3.1

```python
result = generar_receta_y_categorias(ingredients, model = "llama3.1:latest")
print(result)
```

#### Mistral7b

```python

result = generar_receta_y_categorias(ingredients, model = "mistral:latest")
print(result)
```

#### Phi3

```python
result = generar_receta_y_categorias(ingredients, model = "phi3:latest")
print(result)
```

#### TinyLlama

```python
result = generar_receta_y_categorias(ingredients, model = "tinyllama:latest")
print(result)
```

#### Gpt2

```python
result = generar_receta_y_categorias(ingredients, model = "mapler/gpt2:latest")
print(result)
```
